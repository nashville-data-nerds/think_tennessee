```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wages = pd.read_csv('../../data/raw/Median_Wages.csv')
```

```python
wages.head()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
wages.plot(x = 'Year', y = 'Tennessee', ax = ax)
wages.plot(x = 'Year', y = 'United_States', ax = ax, label = 'United States')
ax.set_title('Median Wages Tennessee and US, 1979 - 2018', fontname = 'Lato', fontweight = 700)
plt.sav;
```
